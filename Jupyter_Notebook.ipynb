## Arithmetic Expression Examples

In this section, we will explore various arithmetic expressions using basic mathematical operations. Arithmetic expressions allow us to perform calculations and manipulate numerical values. Here are some examples:

1. Addition: 
   - `2 + 3 = 5`
   - `7.5 + 2.1 = 9.6`
   
2. Subtraction:
   - `8 - 5 = 3`
   - `10.2 - 3.7 = 6.5`
   
3. Multiplication:
   - `4 * 5 = 20`
   - `2.5 * 3.3 = 8.25`
   
4. Division:
   - `10 / 2 = 5`
   - `15.6 / 4 = 3.9`
   
5. Exponentiation:
   - `2 ** 3 = 8`
   - `4 ** 0.5 = 2`
   
6. Modulo (Remainder):
   - `10 % 3 = 1`
   - `15 % 4 = 3`

These examples represent basic arithmetic operations that can be performed in various programming languages, including Python and R. By using these operators, we can perform calculations and solve mathematical problems programmatically.


```python
# Multiplication
a = 5
b = 3
product = a * b
print("The product of", a, "and", b, "is", product)

# Addition
x = 10
y = 7
sum = x + y
print("The sum of", x, "and", y, "is", sum)
```
